<h2 style = "color:red"> Creation of the dictionary.</h2>
This part of the script creates all the possibilities for each card and passes the possibilities to the dictionary called deck. The key for each card is a number between one and 57 and the values are the possibilities of numbers between one and 57 with one possibility common among each card. 

In order to allow for the game to be played numerous times, this code was placed inside a function. which can be invoked inside a while loop so that the dictionary will be continuoulsy replensished within the game.

This function is invoked initialy so that the dictionary's validity can be checked. 

The module random is imported to allow for a random key to be selected from the dictionary.
The module sys is imported to allow for the game to be exited when a certain value is imputted.

```python
# nIm - 1 must be prime
# Cards must have 3, 4, 6 or 8 images
import random
import sys

#create empty dictionary deck that will store cards
deck={}
def create_dictionary():
    """Function that creates the dictionary that stores 57 cards possibilities."""
    nIm = 8
    n = nIm - 1
    r = range(n)
    rp1 = range(n+1)
    c = 0



    # First card
    c += 1
    #deck={}
    value1 = []
    for i in rp1:
        value1.append(i+1)
        deck[c] =value1
 


    # n following cards
    for j in r:
        value2 =[]
        c = c+1
        value2.append(1)
        for k in r:
            value2.append(n+2 + n*j +k)
            
            deck[c] =value2

     


    # n x n following cards
    for i in r:
        for j in r:
            values = []
            c = c+1
            
            values.append(i+2)
            for k in r:
                values.append((n+1 +n*k + (i*k+j) % n)+1)
                deck[c] =values
                
create_dictionary()
```

<h2 style = "color:red">Validity check</h2>
The check validity function tests to see that only one face or number on each card can be found on any other card. In order to test these two for loops are used to create a deck and a duplicate deck, D1 and D2 respectively.  
A variable Inter is created which is the intersection between the two decks.Two options are possible for this function a verbose and silent option.

The verbose option test to see if the intersection of the two cards is greater or less than one and prints a relevant message. If either occurs 1 is added to an empty list l1.
A deck is deemed to be valid if the list does not contain the number 1. The silent option carries out the same test but only prints if the deck is valid or not.


```python
def check_validity(deck, verbose= ""):
    """Function to check validity of deck"""
    l1 =[] 
   # Change the lists in the dictionary to sets
    for k, v in deck.items():
        deck[k] = set(v)

    for i in deck:
        d1 = deck[i]


        for j in deck:
            d2 = deck[j]

        if d1 == d2:
            continue

        #compare the two decks to see that only one image on a card matches
        inter = len(d1.intersection(d2))
        
        if verbose == True:
            #print(inter)
            if inter >1:
                print(str(d1.intersection(d2)),"Too many matches in card" , str(i),"!")
                l1.append(1)
                
            elif inter <1:
                print(str(d1.intersection(d2)),"Too few matches in card" , str(i),"!" )
                l1.append(1)
                
            else:
                print( str(d1.intersection(d2)),"card" , str(i), "is valid")
                
        if verbose == "":
            if inter >1 or inter <1 :
                l1.append(1)
                

                

    if 1 in l1:
        print("deck is invalid")
    else:
        print("\ndeck is valid")
            
            

            
    # change the sets back to lists        
    for k, v in deck.items():
        deck[k] = list(v)
```

<h2 style = "color:red"> import Emojis.</h2>
This section involves selecting a number of emojis and appending them to a dictionary. This dictionary imageDict has a number between 1 and 57 for its key and the value is the emoji.
A search and replace function is used to replace the values in the dictionary deck with the corresponding value in the imageDict.

This code was placed inside function so that it can be invoked within the Dobble game.


```python
def append_emojis():
    """Function to replaces number in dictionary with emojis"""
    import emoji
    imageDict = dict()
    fin = open('emoji_names.txt',"r")
    lines = fin.readlines()
    for i, el in enumerate(lines):
        imageDict[i+1] = emoji.emojize(el.strip())

    #replace the numbers in the deck with corresponding emojis
    for key in deck:
        li = deck[key]
        for i,item in enumerate(li):
            li[i] = imageDict.get(item, item)
            
        
```

<h2 style = "color:red"> dobble Game Classes.</h2>
The Dobble game has two classes within it Dobble card and Dobble deck. When the Dobble deck is instantiated, it initialises itself and creates an empty list called hand.

the Dobble card class initialises itself then randomly chooses an key from the dictionary deck and this key is stored in th variable card. In order to prevent the same card being chosen numerous times from the dictionary, the key value paor associated with the card is removed from the dictionary using the pop function. 

The Dobble deck class has an argument set which represents the number of tests that a user wishes to play.
The Dobble deck class has 3 methods associated with it the add card method which creates an instance of the Dobble card class and appends the selected key to the list hand. 

The remove card function removes the first card from the hand list. 


The play card function selects the first item from the hand list and then returns the values from the dictionary deck which has the same key number as the slelected card.

 



```python
class DobbleDeck():
    def __init__(self,sets):
        self.sets = sets
        self.hand = []
        
        for i in range(sets+1):
            self.add_card()     

                        
    #add the emojis to a dictionary      
    def add_card(self):
        self.hand.append(DobbleCard())
       

    def remove_card(self):
        """Function to remove the first card that has already been played"""
        self.hand.pop(0)
    
    def play_card(self):
        """Function to print the values associated with a sepecific key"""
        c1= self.hand[0].card
        return deck.get(c1)
        
            
class DobbleCard():
    def __init__(self):
        self.card =random.choice(list(deck.keys()))
        deck.pop(self.card)  
```

<h2 style = "color:red"> dobble Game function.</h2>

The first actions executed by the game is to invoke the create dictionary function for the purpose of selecting random keys from the dictionary.

Two variables play1Total and play2Total are assigned the value zero. These variables will store the scores for player one and two.

A while loop is used to read input form the user until an acceptable input is given which is either an integer between 1 and 55 or the string q. If the letter q is typed the ‘sys.exit’ function  is called to end the program. A try statement is used to test whether the input is a string other than q. If all conditions are met then the string ‘set’ is converted into an integer and then used as the argument when creating an instance of the ‘DobbleDeck’ class. If a float is entered The  variable 'a' will remain true thus "Enter valid integer" will be printed.

The create dictionary function is re-invoked in ordre to replenish the deck and alow for the values associated with the keys stored in the variable hand to be displayed.

The append emojis function is called in order to represent the values on the cards as emojis.

A while loop is used to create an instance of the play card function within the DobbleDeck, which is then set as the variable c1. The remove card function is then invoked in order to remove the first item from the hand list. A second instance of the play card is set as the variable c2.

The two variables c1 and c2 are then printed to the screen side by side.

A while loop is also used to read the input that the user types to state which player won or if it was a draw. Each time  they enter a valid value the total points is updated. If a user enters an incorrect value then it will prompt to enter a valid value. If a user enters q the 'sys.exit' function is called to end the program.

The invocation of the Dobble game is placed inside a while loop to allow the game to be played infinitely or until a user enters 'q' or 'Q'.



```python
def dobbleGame():
    """ Dobble game allowing two users to play against one another
    user who find the matching item between both cards must type either 'a' or'b'
    if a draw occurs type 'd'"""
    
    #Call the create dictionary funtcion in order to choose a random card from the dictionary
    create_dictionary()
    
    #dobble scoring system
    play1Total = 0
    play2Total = 0

    # prompt user to enter how many tests they want to play.
    a= True
    while a:
        sets= (input("\nHow many cards (<56)? "))
        print("If you want to record a draw type 'd' or 'D'.")
        print("Enter 'q' or 'Q' to quit")

        # test to see if user enters correct integer for sets or q
        try:
            if sets == "q" or sets =="Q":
                sys.exit()
                
            elif sets != "q" or sets !="Q":
                int(sets)
                if int(sets) >= 56:
                    continue
        except ValueError:
            print("Enter valid integer")
            continue
        

        
        else:
            s = int(sets)
            a = False
           
    #create instance of the dobble game
    play = DobbleDeck(s)
    
    #replenish the deck
    create_dictionary()
    
    #replace text with emojis
    append_emojis()  

    while s > 0:
        c1 = play.play_card()
        play.remove_card()
        c2 = play.play_card()
        
        
        #print emojis side by side
        print(c1[0],c1[1],c1[2], "\t" , c2[0],c2[1],c2[2])
        print(c1[3],c1[4],c1[5], "\t" , c2[3],c2[4],c2[5])
        print(c1[6],c1[7], "\t\t" , c2[6],c2[7])
       
        #ask the user to enter who wins or draws and allow theuser to quit
        flag = True
        while flag is True:
            input_win= (input("\nWho wins (A or B): "))
            if input_win =="a" or input_win =="A":
                play1Total += 1
                s-=1
                
                #print("Score: \nA:" + str(play1Total)+ "\nB:" + str(play2Total))
                flag = False

            elif input_win =="b" or input_win =="B":
                play2Total += 1
                s-=1
                #print("Score: \nA:" + str(play1Total)+ "\nB:" + str(play2Total))
                flag = False

            elif input_win =="d" or input_win =="D":
                play2Total += 1
                play1Total += 1
                s-=1
                
                flag = False
                
            elif input_win =="q" or input_win =="Q":
                sys.exit()
            
            else:
                print("Enter valid key:'a','b','d' or 'q'")
            
        #print score and exit
        if s == False:
            print("Score: \nA:" + str(play1Total)+ "\nB:" + str(play2Total))
        
        
        
```
